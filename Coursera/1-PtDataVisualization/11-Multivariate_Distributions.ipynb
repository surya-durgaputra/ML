## Multivariate Distributions in Python

Sometimes we can get a lot of information about how two variables (or more) relate if we plot them together. This tutorial aims to show how plotting two variables together can give us information that plotting each one separately may miss.



```python
# import the packages we are going to be using
import numpy as np # for getting our distribution
import matplotlib.pyplot as plt # for plotting
import seaborn as sns; sns.set() # For a different plotting theme
%matplotlib inline

# Don't worry so much about what rho is doing here
# Just know if we have a rho of 1 then we will get a perfectly
# upward sloping line, and if we have a rho of -1, we will get 
# a perfectly downward slopping line. A rho of 0 will 
# get us a 'cloud' of points
r = 1

# Don't worry so much about the following three lines of code for now
# this is just getting the data for us to plot
mean = [15, 5]
cov = [[1, r], [r, 1]]
x, y = x, y = np.random.multivariate_normal(mean, cov, 400).T

# Adjust the figure size
plt.figure(figsize=(10,5))

# Plot the histograms of X and Y next to each other
plt.subplot(1,2,1)
plt.hist(x = x, bins = 15)
plt.title("Marginal Distribution of X")

plt.subplot(1,2,2)
plt.hist(x = y, bins = 15)
plt.title("Marginal Distribution of Y")

plt.show()
```

It appears that both of the above distributions don't relate to eachother, other than both having a normal distribution shape. One is centered at 15, the other is centered at 5. We've set the variance of both to be 1 (diagonal of covariance matrix). This means that both of their standard deviations are equal to 1. So we see that they both have the same spread. IF we were to plot these two normally, we will not discern much of association between the two.

```python
# Plot the data
#first set the figure size. We are saying here that plz make my figure 10x10 
plt.figure(figsize=(10,10))
#put this on the top right
plt.subplot(2,2,2)
#scatter command will plot them together
plt.scatter(x = x, y = y)
#lastly set a title
plt.title("Joint Distribution of X and Y")

# Plot the Marginal X Distribution
plt.subplot(2,2,4)
plt.hist(x = x, bins = 15)
plt.title("Marginal Distribution of X")


# Plot the Marginal Y Distribution
plt.subplot(2,2,1)
plt.hist(x = y, orientation = "horizontal", bins = 15)
plt.title("Marginal Distribution of Y")

# Show the plots
plt.show()
```

So in the joint distribution plot, we see that there is a perfect relationship between the the X and Y. And there should be as they are the same data. So we see that sometimes, when we plot two data separately, they might appear unrelated (as inX and Y which were centered around different values) but in fact there was a very strong relationship between the two as shown by the joint plot. Hence joint plots (scatter plots) are very useful in finding such relationships between two variables.
